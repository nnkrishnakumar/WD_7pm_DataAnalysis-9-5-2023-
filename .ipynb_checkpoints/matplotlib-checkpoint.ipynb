```python
from matplotlib import pyplot as plt
```

```python
import random
```

```python
random.randint(1,10)
```

```python
salary=[]
for i in range(20):
    salary.append(random.randint(30000,100000))
```

```python
salary.sort()
print(salary)
```

```python
yearOfExp=[]
for i in range(20):
    yearOfExp.append(random.randint(1,20))
```

```python
yearOfExp.sort()
yearOfExp
```

```python
# line chart/frequency chart
plt.plot(yearOfExp,salary)
plt.show()
```

```python
plt.scatter(yearOfExp,salary)
plt.x
plt.show()
```
