```python
from matplotlib import pyplot as plt
```

```python
import random
```

```python
random.randint(1,10)
```

```python
salary=[]
for i in range(20):
    salary.append(random.randint(30000,100000))
```

```python
salary.sort()
print(salary)
```

```python
yearOfExp=[]
for i in range(20):
    yearOfExp.append(random.randint(1,20))
```

```python
yearOfExp.sort()
yearOfExp
```

```python
# line chart/frequency chart
plt.plot(yearOfExp,salary)
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(yearOfExp,salary,marker="D",edgecolor="red",c="pink")
plt.xlabel("Year of Experience",color="red",size=20)
plt.xticks([i for i in range(0,25)],size=20)
plt.yticks([i for i in range(10000,110000,10000)],size=20)
plt.ylabel("Salary",color="b",size=20)
plt.show()
```
